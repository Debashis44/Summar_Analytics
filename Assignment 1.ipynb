### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
print("\n--- Identifying Consistent Car Models ---")

# Step 0: Ensure 'name' is a column, not the index, and crucial columns are numeric.
# After Task 8, 'hp_per_weight' is the index, and 'name' should be a column.
# If 'name' is still the index for any reason, reset it.
if df.index.name == 'name' or df.index.name == 'hp_per_weight':
    df_reset = df.reset_index()
    print("DataFrame index reset to ensure 'name' is a column.")
else:
    df_reset = df.copy() # Work on a copy if index is already good.

# Ensure 'mpg' and 'model_year' are numeric, handling errors and NaNs
for col in ['mpg', 'model_year']:
    if col in df_reset.columns:
        df_reset[col] = pd.to_numeric(df_reset[col], errors='coerce')
        if df_reset[col].isnull().any():
            if col == 'model_year':
                df_reset[col] = df_reset[col].fillna(df_reset[col].median()) # Use median for year
            else:
                df_reset[col] = df_reset[col].fillna(df_reset[col].mean())
            print(f"Filled NaN in '{col}'.")
    else:
        print(f"Error: Required column '{col}' not found. Cannot proceed with consistency check.")
        # Exit or handle error gracefully if a critical column is missing
        exit() # Exiting as these are critical for the task

# Step 1 & 2: Identify car names that appear in more than one model_year
# and compute the standard deviation of mpg across years, and average mpg, and number of appearances.
if 'name' in df_reset.columns:
    car_model_stats = df_reset.groupby('name').agg(
        num_appearances=('model_year', 'nunique'), # Number of unique years produced
        std_mpg=('mpg', 'std'),                     # Standard deviation of MPG
        avg_mpg=('mpg', 'mean')                     # Average MPG
    )

    # Filter for cars that appear in more than one model_year
    multi_year_models = car_model_stats[car_model_stats['num_appearances'] > 1]
    print(f"\nFound {len(multi_year_models)} car models appearing in more than one year.")

    # Step 3: Filter for standard deviation of mpg < 1.0
    consistent_models = multi_year_models[multi_year_models['std_mpg'] < 1.0]
    print(f"Found {len(consistent_models)} consistent car models (std(mpg) < 1.0).")

    if not consistent_models.empty:
        # Find the car(s) with the lowest variation in mpg among these
        lowest_std_mpg_value = consistent_models['std_mpg'].min()
        lowest_variation_cars = consistent_models[consistent_models['std_mpg'] == lowest_std_mpg_value]

        print(f"\nCar(s) with the lowest variation (std(mpg) = {lowest_std_mpg_value:.4f}) among consistent models:")

        # Step 4 & 5 (Bonus): Report and Sort the result
        # Sort by number of appearances (descending), then mpg (descending)
        final_report = lowest_variation_cars.sort_values(
            by=['num_appearances', 'avg_mpg'],
            ascending=[False, False]
        )

        # Print the required information
        print(final_report[['num_appearances', 'avg_mpg', 'std_mpg']])

    else:
        print("No consistent car models found with standard deviation of mpg less than 1.0.")
else:
    print("Error: 'name' column not found in the DataFrame. Cannot identify consistent car models.")

print("\n--- Consistent Car Model Identification Complete ---")
```
